Given is an implementation of how to scrape content(poem) from a particular webpage(https://www.poetryfoundation.org/poems/browse).
Resources:

- [FCC's vid. tutorial](https://youtu.be/XVv6mJpFOb0?si=eOGBWvwDBDwOHskB)

```python
import requests
from bs4 import BeautifulSoup
import time

def scrape_poems(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        poems = []
        rows = soup.find_all('article')
        if not rows:
            print("No articles found with the specified class.")
            return
        for row in rows:
            title_cell = row.find('a', class_="link-underline-on link-red")
            poet_cell = row.find('div', class_="type-kappa text-gray-600")
            if title_cell and poet_cell:

                title = title_cell.find('span').text.strip()
                poem_url = "https://www.poetryfoundation.org" + title_cell['href']
                poet = poet_cell.find('span').text.strip()

                print(f"Title: {title}, Poem URL: {poem_url}, Poet: {poet}")
                poems.append([title, poem_url, poet])

            # Uncomment the line below during debugging if needed
            # time.sleep(1)
    else:
        print(f"Failed to retrieve page: {response.status_code}")

scrape_poems("https://www.poetryfoundation.org/poems/browse")

```

Your task is to scrape the content(similar) from the *https://poets.org/poems?page=0* and return a code *(with comments explaining what you have done)* and obviously, the scraped material.
- Brownie points for scraping multiple pages (*hint: use "for" loops, and toggle "page=x" in the url, for scraping multiple pages*), along with the poem content


```python
def scrape_poems(page_no):
  page = requests.get(f"https://poets.org/poems", params = {'page':page_no})
  soup = BeautifulSoup(page.content, "html.parser")

  titles = soup.select('td')

  poems = []
  title = author = year = poem_url = None

  for i in range(0, len(titles), 3):  # Iterate in steps of 3 (since each poem has 3 td elements)
      title_td = titles[i]  # Title <td>
      author_td = titles[i+1]  # Author <td>
      year_td = titles[i+2]  # Date <td>

      # Extract title and URL
      if 'views-field-title' in title_td.get('class', []):
          title = title_td.text.strip()
          poem_url = 'https://poets.org' + title_td.select('a')[0].get('href').strip()

      # Extract author
      if 'views-field-field-author' in author_td.get('class', []):
          author = author_td.text.strip()

      # Extract year
      if 'views-field-field-date-published' in year_td.get('class', []):
          year = year_td.text.strip()

      # Append poem details to the list
      poems.append([title, poem_url, author, year])

      # Print details for each poem
      print(f"Title: {title}, Poem URL: {poem_url}, Poet: {author}, Published: {year}")



page_no = int(input("Enter the page No: "))    # which page u want to acess
scrape_poems(page_no)
```
